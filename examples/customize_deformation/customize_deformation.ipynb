# For Points

### Load the example data
- "pts" are points saved as a float32 numpy arrays (N_points,3), two pointsets should share the ordering, i.e. one should be able to find the same neuron by pts[i_neuron] irrelevant of which pointset is loaded
- "frame" are images saved as a float32 numpy arrays (Channel,Width,Height,Depth)
- "mask" are label masks saved as uint8 numpy arrays (Width,Height,Depth)

```python
example_folder="./data/data_points/ex6/"#change the example folder ex1~6
print("Example Folder",example_folder)
pts_tar=np.load(os.path.join(example_folder,"pts_tar.npy"))
frame_tar=np.load(os.path.join(example_folder,"frame_tar.npy"))
pts_gt=np.load(os.path.join(example_folder,"pts_gt.npy"))
frame_gt=np.load(os.path.join(example_folder,"frame_gt.npy"))
mask_gt=np.load(os.path.join(example_folder,"mask_gt.npy"))
C,W,H,D=frame_gt.shape
```

### Perform deformation, this might take up to a minute especially the first time since it sets the GPU ready
Use this part to tune the deformation to your data. Here are the parameters:
- k_cut_dimless: the maximum absolute k of wavevectors to use in units of the fundamental grid frequency
- lr: learning rate of the deformation field.
- iterations: optimization iteration of the defomation field
- frac: the amplitude fraction of the final deformation field used
- lambda_div: the parameter stabilizing the squared divergence of the field
- at_least: minimum number of points required for a succesful deformation


```python
deformparams={"k_cut_dimless":2.5,"lr":0.1,"iterations":1200,"frac":0.3,"lambda_div":1,"at_least":5}
deformation,success=Deformation.get_deformation(name="lowk2d",ptfrom=pts_gt[:,:2],ptto=pts_tar[:,:2],sh=(W,H,D),device=device,**deformparams)
if success:
    with torch.no_grad():
        frame_gt_tr=torch.tensor(frame_gt).to(device=device,dtype=torch.float32)
        mask_gt_tr=torch.tensor(mask_gt).to(device=device,dtype=torch.long)
        frame_def,mask_def=Deformation.deform(frame_gt_tr,deformation,mask=mask_gt_tr)
        frame_def=frame_def.cpu().numpy()
        mask_def=mask_def.cpu().numpy()
print("Success:",success)
```

### Plot the results

```python
colors=np.random.random((mask_gt.max()+1,3));colors[0]=0
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.imshow(frame_gt[0].max(2).T)
plt.scatter(pts_gt[:,0],pts_gt[:,1],c="r",s=1)
plt.title("Original GT Image")
plt.subplot(2,3,2)
plt.imshow(frame_def[0].max(2).T)
plt.title("Deformed GT Image")
plt.subplot(2,3,3)
plt.imshow(frame_tar[0].max(2).T)
plt.scatter(pts_tar[:,0],pts_tar[:,1],c="r",s=1)
plt.title("Target Image")
plt.subplot(2,3,4)
plt.imshow(colors[mask_gt.max(2).T])
plt.title("Original GT Mask")
plt.subplot(2,3,5)
plt.imshow(colors[mask_def.max(2).T])
plt.title("Deformed GT Mask")
plt.subplot(2,3,6)
plt.imshow(frame_def[0].max(2).T-frame_gt[0].max(2).T)
plt.title("Deformed - Original GT Image")
plt.tight_layout()
plt.show()
```

```python
#setup
address_tar  ='./data/data_mask/target' # Target folder
address_gt  ='./data/data_mask/gt' # GT folder
pixel_scale=(0.1625, 0.1625, 1.5) #Pixel Scale
#load some data
target_gt_pairs = [[619,520],[23,620],[22,672]] #We show 3 examples
i_tar,i_gt=target_gt_pairs[0] #try out 0,1,2
frame_tar = np.load(os.path.join(address_tar, "frame_" + str(i_tar) + ".npy"))
mask_tar = np.load(os.path.join(address_tar, "mask_" + str(i_tar) + ".npy"))
frame_gt = np.load(os.path.join(address_gt, "frame_" + str(i_gt) + ".npy"))
mask_gt = np.load(os.path.join(address_gt, "mask_" + str(i_gt) + ".npy"))
```
